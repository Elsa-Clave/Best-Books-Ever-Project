```python
#Standardize header names
cols = []
for i in range(len(best.columns)):
    cols.append(best.columns[i].lower())
best.columns = cols
```

```python
best.info()
```

```python
#Price column: convert type from object to float, and handle issues with the conversion (after encountering errors)
```

```python
#Handle NaN values by filling them with an empty string (otherwise It's not possible to change variable type)
best['price'] = best['price'].fillna('')

#Clean the data
best['price'] = best['price'].str.replace('.', '', regex=False)
best['price'] = best['price'].str.replace(',', '.', regex=False)

#If  there are empty strings  after cleaning, replace them with NaN (will be dealed with after considering each case)
best['price'] = best['price'].replace('', float('nan'))

#Finally Convert to float
best['price'] = best['price'].astype(float)
```

```python
#Check if the result is as expected:
print(best['price'].dtype)
```

```python
#Pages column: convert to integer, handle errors encountered such as non-numeric characters present in some observations.
```

```python
#Remove non-numeric characters with regex
best['pages'] = best['pages'].str.extract('(\d+)', expand=False)

#As before, fill NaN values with a default value. (`0`values will be dealed with after considering each case)
best['pages'] = best['pages'].fillna(0)

#Convert the column to integers
best['pages'] = best['pages'].astype(int)
```

```python
#Check if the result is as expected:
print(best['pages'].dtype)
```

```python
#Publishdate and firstpublishdate columns: from object (string) format to datetime format:
```

```python
# Check problematic entries for 'publishdate'
problematic_publishdate = best[~best['publishdate'].apply(pd.to_datetime, errors='coerce').notna()]
print("Problematic publishdate entries:")
print(problematic_publishdate)
```

```python
#Replacing the problematic entries with NaT (missing)
best['publishdate'] = best['publishdate'].apply(lambda x: pd.NaT if pd.to_datetime(x, errors='coerce') is pd.NaT else x)
```

```python
#Convert columns to datetime format, handling errors with 'coerce'
best['publishdate'] = pd.to_datetime(best['publishdate'], errors='coerce')
```

```python
#Check if the result is as expected:
print(best['publishdate'].dtype)
```

```python
#Same process to checj problematic entries this time for 'firstpublishdate'
problematic_firstpublishdate = best[~best['firstpublishdate'].apply(pd.to_datetime, errors='coerce').notna()]
print("Problematic firstpublishdate entries:")
print(problematic_firstpublishdate)
```

```python
#Replacing the problematic entries with NaT (missing)
best['firstpublishdate'] = best['firstpublishdate'].apply(lambda x: pd.NaT if pd.to_datetime(x, errors='coerce') is pd.NaT else x)
```

```python
#Convert columns to datetime format, handling errors with 'coerce'
best['firstpublishdate'] = pd.to_datetime(best['firstpublishdate'], errors='coerce')
```

```python
#Check if the result is as expected:
print(best['firstpublishdate'].dtype)
```

```python
#The column ratingbystars indicates the ratings for 5,4,3,2 and 1 stars like so ['1593642', '637516', '222366', '39573', '14526']
#The idea is to create separated columns for better clarity (more understandable)
#Also for better analysis(avg rating, total ratings, distribution of ratings)
```

```python
#First (after many tests) let's make sure all values are treated as strings and replace None with '[]' 
best['ratingsbystars'] = best['ratingsbystars'].apply(lambda x: str(x) if x is not None else '[]')

#Then remove the square brackets and split the string into a list of ratings
best['ratingsbystars'] = best['ratingsbystars'].str.strip('[]').str.replace("'", "").str.split(',')

#Changing empty strings or None with '0'
best['ratingsbystars'] = best['ratingsbystars'].apply(lambda x: ['0' if elem.strip() == '' else elem.strip() for elem in x])

#Each list has to have 5 elements (5 ratings)
best['ratingsbystars'] = best['ratingsbystars'].apply(lambda x: x + ['0'] * (5 - len(x)) if len(x) < 5 else x)

#Creating new columns
ratings_df = pd.DataFrame(best['ratingsbystars'].tolist(), index=best.index, columns=['5_star', '4_star', '3_star', '2_star', '1_star'])

#Changing type!
ratings_df = ratings_df.astype(int)

#Merge the new columns back
best = best.drop(columns=['ratingsbystars'])
best = pd.concat([best, ratings_df], axis=1)


print(best.dtypes)
```

```python
def calculate_null_percentage(data):
    null_percentage = round(data.isna().sum() / len(data) * 100, 4)
    nulls_df = pd.DataFrame(null_percentage).reset_index()
    nulls_df.columns = ['header_name', 'percent_nulls']
    return nulls_df
```

About some variables nulls:
- `series `: Since null values indicate that the book is not part of any series, the idea is to replace nulls with a placeholder 
- `description, language, bookFormat, publisher`: replacing null values with a placeholder such as "Unknown" to ensure consistency while acknowledging missing data
- `edition`: given the low number of non-null values the best may be to keep them as this data is not critical. (For now)
- `publishDate, firstPublishDate`: as Nulls in date columns can be tricky. I might fill nulls with a placeholder date such as pd.NaT or None
- `likedPercent, coverImg, price`: For this numerical columns like likedpercent and price, I might replace values with a central tendency measure such as the mean or median ans for coverimg, replace nulls with "No cover available"

```python
#Replacing nulls in 'series' with 'Not part of a series'
best['series'].fillna('Not part of a series', inplace=True)
```

```python
# Replacing nulls in 'description', 'language', 'bookformat', 'publisher' with 'Unknown'
best['description'].fillna('Unknown', inplace=True)
best['language'].fillna('Unknown', inplace=True)
best['bookformat'].fillna('Unknown', inplace=True)
best['publisher'].fillna('Unknown', inplace=True)
```

```python
#Replacing nulls in 'edition' with 'Unknown'
best['edition'].fillna('Unknown', inplace=True)
```

```python
#Replacing nulls in 'publishdate' and 'firstpublishdate' with pd.NaT (Not-a-Time)
best['publishdate'].fillna(pd.NaT, inplace=True)
best['firstpublishdate'].fillna(pd.NaT, inplace=True)
```

```python
#Replacing nulls in 'likedpercent' with the mean of the column
best['likedpercent'].fillna(best['likedpercent'].mean(), inplace=True)
```

```python
#Replacing nulls in 'coverimg' with 'No cover available'
best['coverimg'].fillna('No cover available', inplace=True)
```

```python
#Replacing nulls in 'price' with 0 (for now)
best['price'].fillna(0, inplace=True)
```
